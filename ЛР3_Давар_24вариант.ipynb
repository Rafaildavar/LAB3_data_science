## Первая часть - простая линейная регрессия:


1-2. Необходимо обучить модель простой (парной) линейной регрессии, используя для обучения значения x1 и y из таблицы по варианту и сделать предсказание.

x1 = 3, 3, 6, 6, 7, 8, 9;

y = 26.5, 26.4, 28.2, 27.6, 26.9, 25.2, 26.6;

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

data = {
    'x1': [3, 3, 6, 6, 7, 8, 9],
    'target': [26.5, 26.4, 28.2, 27.6, 26.9, 25.2, 26.6]
}
df = pd.DataFrame(data)

y = df['target']
X = df[['x1']]

model = LinearRegression()
model.fit(np.array(X).reshape(-1,1), y)
predictions_valid = model.predict(np.array(X).reshape(-1,1))
```

3-4. После этого нужно создать и вывести датафрейм с истинными значениями и предсказанными Также нужно подсчитать и вывести метрики качества регресии.

```python
from sklearn import metrics
df_match = pd.DataFrame({'Actual': y, 'Predicted': predictions_valid})
print('Mean Absolute Error:', metrics.mean_absolute_error(y, predictions_valid))
print('Mean Squared Error:', metrics.mean_squared_error(y, predictions_valid))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y, predictions_valid)))
print('R2 metric:', metrics.r2_score(y, predictions_valid))
df_match
```

* Средняя абсолютная ошибка составляет примерно 0.69.
Это значение показывает среднюю разницу (по модулю) между фактическими значениями и предсказанными. MAE довольно близко к нулю, что должно указывать на точность модель.
* Среднеквадратическая ошибка равна 0.76. Это метрика, чувствительная к крупным ошибкам, так как она возводит в квадрат отклонения. Поскольку она больше, чем MAE, это указывает на наличие больших отклонений в некоторых предсказанных значениях.
* Корень из среднеквадратичной ошибки равен 0.87. Это значение близко к MSE, но интерпретируется в тех же единицах, что и данные, как и MAE.
* Значение R² показывает какую долю дисперсии зависимой переменной объясняет модель. Это значение около 0.015, что указывает на то, что модель объясняет около 1.5% вариации в целевой переменной. Это означает, что модель объясняет очень небольшую часть вариации в данных.


5-6. Далее нужно вывести значения коэффициентов a и b. После необходимо выполнить визуализацию регрессии: точки (scatter plot) и линию регрессии.

```python
import matplotlib.pyplot as plt

a = model.coef_
b = model. intercept_

print(a,b)
plt.scatter(df['x1'], df['target'], label = 'Исходные данные')

plt.plot([3, 9], [a*3 + b, a*9 + b], "r", label='Линия регрессии')
plt.title('Линейная регрессия')
plt.legend()

plt.show()
```

С помощью метода scatter и plot были выведены точки и линия регрессии на основе предсказанных данных

7. Далее необходимо построить график с разницей предсказанного и истинного значения по каждой точке.

```python
from yellowbrick.regressor import ResidualsPlot
visualizer = ResidualsPlot(model)
visualizer.fit(X, y)
visualizer.score(X, y)
visualizer.show()
```

По данному графику можно сделать вывод, что разница между наблюдаемыми и прогнозируемыми значениями находится в диапазоне от -1.5 до 1.5. Многие точки лежат далеко от параллели, значит модель неточно предсказывает значения.

8. Следует построить еще один график.

```python
plt.figure(figsize=(10, 6))

plt.scatter(y, predictions_valid, color='blue', alpha=0.5, label='Предсказанные значения')
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linewidth=2, label='Фактические значения')
plt.legend()
plt.show()
```

Синяя точечная диаграмма показывает, как предсказанные значения (predictions_valid) соотносятся с истинными значениями (y). Красная линия показывает линию идеальных предсказаний, то есть случаев, когда предсказанное значение точно совпадает с истинным. Значение R2 очень близко к нулю, что подтверждается тем, что точки предсказаний обладают значениями близкими к среднему от истинных значений. Модель неточная и очень похожа на прогноз по среднему значению.

# Часть 2 - Полиномиальная регрессия¶

1-2. Использовать PolynomialFeatures для реализации модели полиномиальной регрессии и обучить модель полиномиальной регрессии, используя данные по варианту.

x = 0, 1, 2, 3, 4, 5, 6;

y = 25.2, 46.2, 56.7, 77.6, 91.5, 112.3, 106.2;

Сначала была выбрана 2я степень полинома для модели полиномиальной регрессии.

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
data2 = {
    'x': [0, 1, 2, 3, 4, 5, 6],
    'target': [25.2, 46.2, 56.7, 77.6, 91.5, 112.3, 106.2]
}

df2 = pd.DataFrame(data2)

y2 = df2['target']
X2 = df2[['x']]

poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(X2)
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y2)
lin_reg.intercept_, lin_reg.coef_
```

3-4. Далее было выполнено предсказание и высчитаны метрики качества регрессии (MAE, R2).

```python
predictions_2_model = lin_reg.predict(X_poly)
df_match_2 = pd.DataFrame({'Actual': y2, 'Predicted': predictions_2_model})
print('Mean Absolute Error:', metrics.mean_absolute_error(y2, predictions_2_model))
print('R2 metric:', metrics.r2_score(y2, predictions_2_model))
df_match_2
```

Средняя абсолютная ошибка модели низкая, что указывает на хорошее приближение предсказаний к реальным значениям. Модель объясняет 97.48% дисперсии данных, что свидетельствует о высоком качестве предсказаний. Можно сказать, что модель хорошо описывает зависимость данных.

5. Выполнение визуализации регрессии: точки и линию регрессии.

```python
plt.scatter(df2['x'], df2['target'], label='Исходные данные')

x_range = np.linspace(df2['x'].min(), df2['x'].max(), 100).reshape(-1, 1)
x_range_poly = poly_features.transform(x_range)
y_range = lin_reg.predict(x_range_poly)

plt.plot(x_range, y_range, color='red', label='Полиномиальная регрессия')
plt.title('Полиномиальная регрессия во 2 степени')
plt.legend()
plt.show()
```

Визуализация подтверждает, что модель очень хорошо предсказывает результаты.

6. Необходимо повторить все пункты 1-5 для еще одной степени полинома (degree). Была выбрана 4я степень полинома для ещё одной регрессии.

```python
poly_features = PolynomialFeatures(degree=4)
X_poly = poly_features.fit_transform(X2)
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y2)
lin_reg.intercept_, lin_reg.coef_

predictions_4_model = lin_reg.predict(X_poly)
df_match_4 = pd.DataFrame({'Actual': y2, 'Predicted': predictions_4_model})
print('Mean Absolute Error:', metrics.mean_absolute_error(y2, predictions_4_model))
print('R2 metric:', metrics.r2_score(y2, predictions_4_model))
df_match_4
```

Средняя абсолютная ошибка модели стала ещё ниже, а модель теперь объясняет 99.60% дисперсии данных, что свидетельствует о ещё большем повышении качества предсказаний.

```python
plt.scatter(df2['x'], df2['target'], label='Исходные данные')

x_range = np.linspace(df2['x'].min(), df2['x'].max(), 100).reshape(-1, 1)
x_range_poly = poly_features.transform(x_range)
y_range = lin_reg.predict(x_range_poly)

plt.plot(x_range, y_range, color='red', label='Полиномиальная регрессия')
plt.title('Полиномиальная регрессия в 4 степени')
plt.legend()
plt.show()
```


По графику и полученным значениям видно, что модель объясняет больше дисперсии данных и становится более точной при возрастании степени полинома. У данной модели значение R² близкое к 1 указывает на то, что модель успешно захватывает зависимость между признаками и целевой переменной в обучающей выборке.


Вывод по второй части:

Исходя из значений метрик, можно сделать вывод, что для данных, полученных для обучения, модель для 4й и более степени обладает очень большой точностью. Чем больше степень полинома, тем коэффициент детерминации всё ближе к 1. Данный результат достигается за счёт переобучения модели, ведь она становится слишком сложной и слишком много подстаивается под обучающие данные, что приведет к плохой обобщающей способности модели на новые данные.

4-7. Необходимо разделить данные на валидационные и обучающиеся выборки. Нормализовать числовые данные с помощью StandardScaler. Обучить модель линейной регрессии с помощью LinearRegression и применить обученную модель на тестовой выборке.

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import Ridge
from sklearn.preprocessing import OrdinalEncoder

ordinal_cols = ['fueltype', 'aspiration', 'carbody', 'drivewheel', 'enginelocation', 'enginetype', 'cylindernumber', 'fuelsystem','carname', 'doornumber']  # Примеры категорий с порядком
ordinal_encoder = OrdinalEncoder()

df3[ordinal_cols] = ordinal_encoder.fit_transform(df3[ordinal_cols])

X3 = df3.drop(columns=['price', 'carID'])
y3 = df3['price']

X3_train,X3_valid,y3_train,y3_valid = train_test_split(X3, y3, test_size=0.25, random_state=0)

scaler = StandardScaler()
X3_train_scaled = scaler.fit_transform(X3_train)
X3_valid_scaled = scaler.transform(X3_valid)

model3 = LinearRegression()
model3.fit(X3_train_scaled, y3_train)
predictions_valid3 = model3.predict(X3_valid_scaled)

print(predictions_valid3[:15])
```

7. Нужно оценить качество модели с помощью метрики.

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y3_valid, predictions_valid3))
print('Mean Squared Error:', metrics.mean_squared_error(y3_valid, predictions_valid3))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y3_valid, predictions_valid3)))
print('R2 metric:', metrics.r2_score(y3_valid, predictions_valid3))
```

Значение R² указывает на то, что примерно 83.1% вариации целевой переменной объясняется моделью, что является хорошим результатом. Исхоя из масштаба данных можно сказать, что показатели метрик MAE, MSE, RMSE являются удовлетварительными.

11. Необходимо реализовать регрессию методом k-ближайших соседей или деревом решений. В данном примере используется дерево решений

```python
from sklearn.tree import DecisionTreeRegressor

dt_model = DecisionTreeRegressor(random_state=0)
dt_model.fit(X3_train_scaled, y3_train)

dt_predictions = dt_model.predict(X3_valid_scaled)

dt_mae = metrics.mean_absolute_error(y3_valid, dt_predictions)
dt_mse = metrics.mean_squared_error(y3_valid, dt_predictions)
dt_rmse = np.sqrt(metrics.mean_squared_error(y3_valid, dt_predictions))
dt_r2 = metrics.r2_score(y3_valid, dt_predictions)

print("Метрики регрессии дерева решений:")
print("Mean Absolute Error:", dt_mae)
print("Mean Squared Error:", dt_mse)
print("Root Mean Squared Error:", dt_rmse)
print("R2 Score:", dt_r2)
```

Как видно из метрик, почти все показатели превосходят показатели линейной регрессии. MAE, MSE, RMSE уменьшили своё значение отклонения, а R² показывает возрастание с 83.1% до 87.74%.
